# Regression 迴歸分析

```python
%matplotlib inline
# %env KERAS_BACKEND=tensorflow
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
# import sympy as sym
# from sympy import S,N
# import locale
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## LinearRegression語法測試

```python
x=np.linspace(0,10,50)
nosie=np.random.randn(50)*10
y=3*x+7+nosie
plt.plot(x,3*x+7,'b');
plt.scatter(x,y);
```

```python
regr=LinearRegression()
X=x.reshape(len(x),1)#必須讓他從[1,1,1,...]的形式變成[[1],[1],[1],...]
regr.fit(X,y)#執行完後就訓練完畢了
```

```python
testing_points=np.array(range(1,8)).reshape(7,1)#注意要reshape
Y_predict=regr.predict(testing_points)
plt.plot(x,regr.predict(X),'r');
plt.plot(x,3*x+7,'b');
plt.scatter(x,y);
plt.scatter(testing_points,Y_predict);
```

## 真正拿資料來時會用的做法

```python
x=np.linspace(0,5,200)
nosie=np.random.randn(200)*10
y=3*x+7+nosie
plt.plot(x,3*x+7,'b');
plt.scatter(x,y);
```

```python
X=x.reshape(len(x),1)
x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.3,random_state=9487)
#random_state=9487<---可以直接空白，這是指定亂數種子
#這代表我們使用70%用來訓練30%用來測試訓練成果
plt.scatter(x_train,y_train);
```

```python
regr=LinearRegression()
regr.fit(x_train,y_train)
```

```python
# plt.plot(x,regr.predict(X),'r');
plt.plot(x,3*x+7,'b');
plt.scatter(x_train,y_train);
Y_predict=regr.predict(x_test)
plt.plot(x_test,Y_predict,'r');
```
