# 可能的錯誤訊息註記

Pandas copy常會出現一種警告訊息"SettingWithCopyWarning"

使用view只是觀看，copy才會真的複製

所以如果直接指定view值，會直接更動到原本的資料!!

而copy不會。


```python
print(男女計.loc[0][0])
男女計.loc[0][0]="政治"
print(男女計.loc[0][0])
print(undergraduate.loc[0][1])#會更動到undergraduate!!所以我們需要使用copy!!
print(df.loc[0][1])#但不會更動到df
print("-------------------------")
print(男女計.loc[0][0])
男女計.loc[0][0]="國立政治大學"
print(男女計.loc[0][0])
print(undergraduate.loc[0][1])#先改回來
print(df.loc[0][1])
```

```python
男女計=undergraduate.loc[:,"學校名稱":"女生計"].copy()#重令使用copy
男女計.head()
```

```python
print(男女計.loc[0][0])
男女計.loc[0][0]="政治"
print(男女計.loc[0][0])
print(undergraduate.loc[0][1])#有copy就不會更動到undergraduate!!
print(df.loc[0][1])#也不會更動到df
男女計.loc[0][0]="國立政治大學"
print(男女計.loc[0][0])
```

```python
import locale
```

```python
locale.setlocale(locale.LC_NUMERIC,"")#回傳目前系統設置的LC_NUMERIC
#!!!必要!!!!
```

```python
男女計.男生計=男女計.男生計.apply(locale.atof)# atof-->to float, atoi-->to int
男女計.女生計=男女計.女生計.apply(locale.atof)#一種技巧將逗號去除，另外可以參考"astype"
```

```python
男女計.dtypes
#成功變成float
```

```python
男女計["男女比"]=男女計.男生計/男女計.女生計
男女計.head()
```

```python
sort_男女計=男女計.sort_values(by=['男女比'],ascending=False).copy()#by=['男女比',男生計]表示排序的優先後
sort_男女計.head()
```

```python
sort_男女計[0:5].plot.bar(x="學校名稱",y="男女比",figsize=(8, 5),fontsize=20);
```
